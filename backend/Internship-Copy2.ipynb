```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = df.drop(['local_time','time'],axis=1)
```

```python
x = df1.drop('electricity', axis=1)
y = df1['electricity']
```

```python
x[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_test[:5]
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
df_test = pd.read_csv("D:/Downloads/ninja_pv_29.4345_3.4351_uncorrected.csv")
df_test.head() 
df_test = df_test.drop(['local_time','time'],axis=1)
```

```python
nx = df_test.drop('electricity', axis=1)
ny = df_test['electricity']
ny_pred = model.predict(nx)
print("MAE:", mean_absolute_error(ny, ny_pred))
print("R² Score:", r2_score(ny, ny_pred))
```
